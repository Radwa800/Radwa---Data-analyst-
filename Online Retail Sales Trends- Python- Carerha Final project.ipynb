```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd. read_csv('/content/drive/MyDrive/Final Project - Sales Data /sales_data_sample.csv', encoding='latin-1')
print(df.dtypes)

# Data Transformation and Cleaning

## Changing Data type
df['ORDERDATE']=  pd.to_datetime=(df['ORDERDATE'])

## Delete Address line 2 Column
df = df.drop('ADDRESSLINE2', axis=1)

## Delete STATE Column
df= df.drop('STATE', axis=1)

## Adding Contact first name and Last name
df['ContactName'] = df['CONTACTFIRSTNAME'] + df['CONTACTLASTNAME']

## Delete Contact First & last name

df = df.drop('CONTACTFIRSTNAME', axis=1)

df = df.drop('CONTACTLASTNAME', axis=1)

df.head(10)


# Aggregate customer data
customer_data = df.groupby('CUSTOMERNAME').agg({
    'SALES': 'sum',  # Total spending
    'ORDERNUMBER': 'count'  # Purchase frequency
}).rename(columns={'SALES': 'TotalSpending', 'ORDERNUMBER': 'PurchaseFrequency'})

# Define thresholds for segmentation
high_value_threshold = customer_data['TotalSpending'].quantile(0.75)
low_value_threshold = customer_data['TotalSpending'].quantile(0.25)

# Assign segments
customer_data['Segment'] = pd.cut(customer_data['TotalSpending'],
                                  bins=[0, low_value_threshold, high_value_threshold, float('inf')],
                                  labels=['Low-Value', 'Mid-Value', 'High-Value'])

# Display segmented customer data
print(customer_data.head())



# Total Revenue per Customer
customer_revenue = df.groupby('CUSTOMERNAME')['SALES'].sum().reset_index()
customer_revenue.columns = ['CUSTOMERNAME', 'total_revenue']

# Average Order Value
average_order_value = df.groupby('CUSTOMERNAME')['SALES'].mean().reset_index()
average_order_value.columns = ['CUSTOMERNAME', 'avg_order_value']

# Purchase Frequency
purchase_frequency = df.groupby('CUSTOMERNAME').size().reset_index(name='SALES')

# Merge data
customer_data = customer_revenue.merge(average_order_value, on='CUSTOMERNAME')
customer_data = customer_data.merge(purchase_frequency, on='CUSTOMERNAME')

# Assume an average churn rate (e.g., 0.2)
churn_rate = 0.2
# Assume a gross margin rate (e.g., 0.50 or 50%)
gross_margin_rate = 0.50
## Calculate Customer life time value

customer_data['cltv'] = (customer_data['avg_order_value'] * customer_data['SALES'] / churn_rate) * gross_margin_rate

df.to_csv('/content/drive/MyDrive/Final Project - Sales Data /sales_data_sample.csv', index=False)
```
